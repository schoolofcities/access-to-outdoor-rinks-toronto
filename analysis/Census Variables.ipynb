## Download and Clean Census Data by CT

```python
import pandas as pd
import numpy as np
```

```python
#Clean CT Data Table
ct = pd.read_csv("data/census/CT/ct_census.csv")

ct = ct.rename(columns = 
              {'v_CA21_560: Median total income in 2020 among recipients ($)':"Median_Income",
               'v_CA21_4410: Immigrants':"Immigrant#",
               "v_CA21_4875: Total visible minority population":"VM#",
               "v_CA21_1025: In low income based on the Low-income measure, after tax (LIM-AT)":"LIncome#",
               'Population ':"Population"
              })

ct.columns = ct.columns.str.replace(" ", "")
```

```python
#Calculate immigrant and visible miniority percentages
ct["Immigrant%"] = ct.apply(lambda x: x["Immigrant#"]/x["Population"]*100, axis=1)
ct["Visible_Minority%"] = ct.apply(lambda x: x["VM#"]/x["Population"]*100, axis =1)
ct["Low-Income%"]=ct.apply(lambda x:x["LIncome#"]/x["Population"]*100, axis=1)

#Calculate population density, immigrant density, visible minority density
ct["Population_Density"] = ct.apply(lambda x: x["Population"]/x["Area(sqkm)"], axis =1)
ct["Immigrant_Density"] = ct.apply(lambda x: x["Immigrant#"]/x["Area(sqkm)"], axis =1)
ct["VM_Density"]= ct.apply(lambda x: x["VM#"]/x["Area(sqkm)"], axis =1)
```

```python
ct["GeoUID"] = ct["GeoUID"].apply(lambda x:"{:.2f}".format(x))
ct
```
